```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt 

# load the image
img = Image.open("meme_4.PNG")

# create  a draw object 
draw = ImageDraw.Draw(img)

# Prepare text
text = "** Ohh Shit !!! ** "

# chose a font and size(using built-infont)
font = ImageFont.load_default()


# add text to the image
draw.text((100,100), text = text, fill = (255,0,0), align = 'Right' ,font = ImageFont.truetype('arial.ttf', 80) )

text = " Ye Kya Hua ??  "
draw.text((400,200), text = text, fill = (0,0,255), align = 'Right' ,font = ImageFont.truetype('arial.ttf', 80) )

# display the image
plt.imshow(img)
plt.show()
```
